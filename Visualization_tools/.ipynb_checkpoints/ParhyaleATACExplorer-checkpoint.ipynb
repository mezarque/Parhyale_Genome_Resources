# 0. Setup
Running the cell below imports all of the necessary packages and sets important functions for downstream plotting.

```python
################################
### Do not modify this block ###
################################

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
pd.options.display.max_colwidth = 100
from IPython.display import display
import os
import warnings
import numpy as np
from scipy import interpolate
from scipy import signal
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

new_rc_params = {'text.usetex': False, "svg.fonttype": 'none'}
plt.rcParams.update(new_rc_params)

sns.set_style({'font.family':'sans-serif', 'font.sans-serif':'Arial'})
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"

def ImpulseFunction(t, vecImpulseParam):
    return (1/vecImpulseParam[3]) * (vecImpulseParam[2] + (vecImpulseParam[3] - vecImpulseParam[2]) * (1/(1 + np.exp(-vecImpulseParam[1] * (t - vecImpulseParam[5]))))) * (vecImpulseParam[4] + (vecImpulseParam[3] - vecImpulseParam[4]) * (1/(1 + np.exp(vecImpulseParam[1] * (t - vecImpulseParam[6])))))

def ImpulseEvaluate(vecTimepoints, vecImpulseParam):
    vecImpulseValue = [ImpulseFunction(i, vecImpulseParam) for i in vecTimepoints]
    vecImpulseOutput = [i if i > 10^(-10) else 10^(-10) for i in vecImpulseValue] 
    return vecImpulseOutput

def get_last(string):
    dummy_array = string.split(';')
    if 'eggNOG-name' in dummy_array[-1].split('='):
        return dummy_array[-1].replace('eggNOG-name=', '')
    else:
        return dummy_array[-1]
    
def igv_to_peak(string):
    contig_contig = string.split(':')[0]
    contig_coords = string.split(':')[1]
    contig_start = contig_coords.split('-')[0].replace(',', '')
    contig_end = contig_coords.split('-')[1].replace(',', '')
    contig = contig_contig + '_' + str(int(contig_start) - 1) + '_' + contig_end
    return contig

#taken from a Stack Overflow post
#https://stackoverflow.com/questions/2566412/find-nearest-value-in-numpy-array
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def chrom_dict_generator(lst):
    dict_dummy = {}
    for item in lst:
        chrom = item.split(':')[0]
        coords = item.split(':')[1]
        cleaned = coords.replace(',', '')
        start = int(cleaned.split('-')[0])
        end = int(cleaned.split('-')[1])
        dict_dummy[chrom] = (start, end)
    return dict_dummy

def plot_genome_region(chrom_sizes, regions_dict, chrom_dict, alias_dict,
                       draw_plots = False, padding = 0, figsize = (5, 3), alpha = 0.5,
                      savefig = False, prefix = '', imgtype = 'png'):
    
    #Load in chrom_sizes file
    chrom_sizes_df = pd.DataFrame(pd.read_csv(chrom_sizes, sep = '\t', header = None))
    chrom_sizes_df[2] = 1
    chrom_sizes_df = chrom_sizes_df[[0, 2, 1]]
    
    #load in chromosome and region specified in chrom_dict
    chrom_sizes_chroms = chrom_sizes_df[chrom_sizes_df[0].isin(chrom_dict.keys())]
    
    #load in data from each region_dict file
    region_data = {}
    
    for j in regions_dict:
        region_data[j] = pd.DataFrame(pd.read_csv(j, sep = '\t', header = None, skiprows = 2))
    
    if draw_plots:
    
        #make a plot for each chromosome and region
        for i in chrom_dict:
            start = chrom_dict[i][0]
            stop = chrom_dict[i][1]
            
            region_name = i + ':'+ str(start) + 'â€“' + str(stop)
            
            plt.figure(figsize = figsize)
            
            #Set line representing baseline genome
            plt.hlines(-1, start, stop, color = 'k', linewidth = 25)
            
            hlines_sofar = 0
            hlines_legend = {}
            
            for j in region_data:
                data = region_data[j]
                overlapping = data[data[0] == i]
                overlapping = overlapping[(overlapping[3] >= start) | (overlapping[4] <= stop)]
                
                for datatype in regions_dict[j]:
                    correct_type = overlapping[overlapping[1] == datatype]
                    
                    if datatype == 'any':
                        correct_type = overlapping
                        
                    hlines_start = correct_type[3] - padding
                    hlines_end = correct_type[4] + padding
                    hlines_y = [hlines_sofar for h in hlines_start]
                    
                    plt.hlines(hlines_y, hlines_start, hlines_end, color = regions_dict[j][datatype], linewidth = 15, alpha = alpha)
                    
                    hlines_sofar += 1
                    
                    hlines_legend[' '.join([alias_dict[j], datatype])] = regions_dict[j][datatype]
                
                plt.hlines(hlines_sofar - 0.5, start, stop, linewidth = 0.5, linestyle = 'dotted', color = 'gray')
                    
            #clear yticks
            plt.yticks([i for i in np.arange(0, len(hlines_legend.keys()))], [i.replace('_', ' ') for i in hlines_legend.keys()])
        
            #set ylims based on number of datatypes being plotted
            plt.ylim(-1, hlines_sofar - 0.5)
        
            ax = plt.gca()
            ax.get_xaxis().set_major_formatter(
                mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
            plt.xticks(rotation = 22.5, ha = 'right', rotation_mode = 'anchor')
            plt.xlabel('genomic coordinate')
            plt.title(region_name)
        
            plt.xlim(start, stop)
            
            if savefig:
                plt.savefig(prefix + '_' + region_name.replace(':', '_') + '.' + imgtype)
            
            plt.show()
    
    return None

def plot_IDE2_models(contig_string, multicovloc, allparamsloc, noS21S22loc,
                     alldata = True, noS21S22data = True, scatter = True, highlightS21S22 = True,
                     figsize = (8, 6), savefig = False, nickname = '', imgtype = 'svg'):
    
    #load in the base data
    contig = igv_to_peak(contig_string)
    ImpulseParams_all = pd.DataFrame(pd.read_csv(allparamsloc))
    ImpulseParams_noS21S22 = pd.DataFrame(pd.read_csv(noS21S22loc))
    multicov = pd.DataFrame(pd.read_csv(multicovloc, sep = '\t'))

    #Extract the parameters for your line of interest
    impulse_params_for_contig_all = ImpulseParams_all.loc[ImpulseParams_all['contig_start_end'] == contig].drop(columns = ['Unnamed: 0']).values.tolist()[0]
    impulse_params_for_contig_noS21S22 = ImpulseParams_noS21S22.loc[ImpulseParams_noS21S22['contig_start_end'] == contig].drop(columns = ['Unnamed: 0']).values.tolist()[0]
    all_multicov = multicov.loc[multicov['contig_start_end'] == contig].drop(columns = ['contig_start_end']).astype('float').values.tolist()[0]
    
    #Calculate fit given Impulse function expected ranges
    x_impulse_linspace = np.linspace(72, 192, 300)
    x_impulse = [72, 77, 80, 87, 90, 96, 104, 112, 120, 132, 144, 155, 168, 180, 192]
    x_impulse = [i for i in x_impulse for _ in (0, 1)]
    impulse_fit_all = ImpulseEvaluate(x_impulse_linspace, impulse_params_for_contig_all)
    impulse_fit_noS21S22 = ImpulseEvaluate(x_impulse_linspace, impulse_params_for_contig_noS21S22)
    
    #Plot lines
    fig, ax = plt.subplots(figsize=figsize)
    plt.title(nickname + ' : ' + contig if nickname != '' else contig)
    
    if alldata:
        plt.plot(x_impulse_linspace, impulse_fit_all, c = 'tab:blue', label = 'IDE2 alldata model fit')
    if noS21S22data:
        plt.plot(x_impulse_linspace, impulse_fit_noS21S22, c = 'tab:orange', label = 'IDE2 noS21S22 model fit')
    if scatter:
        plt.scatter(x_impulse, all_multicov, c = 'gray', label = 'Omni-ATAC raw accessibility')
    if highlightS21S22:
        plt.scatter(x_impulse[16:20], all_multicov[16:20], c = 'tab:orange', label = 'S21-S22 points')
    plt.ylabel('accessibility model fit')
    plt.xlabel('developmental stage')
    plt.yticks(fontsize = 12)
    plt.xticks([72, 77, 80, 87, 90, 96, 104, 112, 120, 132, 144, 155, 168, 180, 192], 
           ['S13', 'S14', 'S15', 'S17', 'S18', 'S19', 'S19+', 'S20', 'S21', 'S22', 'S23', 'S24', 'S25', 'S26', 'S27'], 
           rotation = 'vertical', fontsize = 12)
    plt.legend(loc = 'best')
    
    ax2range = [i for i in range(60, 220, 20)]
    ax2 = plt.twiny()
    ax2.set_xticks([(i - 60) / (200 - 60) for i in ax2range])
    ax2.set_xticklabels(ax2range)
    ax2.set_xlabel('hours post fertilization')
    
    #contig = 'phaw_50.283864a_3320701_3323202'
    plot_name = nickname.replace(' ', '_') + '_' + contig if nickname != '' else contig
    file_name = plot_name + '.' + imgtype
    
    if savefig:
        plt.savefig(file_name)

    plt.show()
    
    return None

def plot_IDE2_region(contig_string, peaksloc, multicovloc, allparamsloc, 
                     alldata = True, scatter = True, acc_filter = 15,
                     figsize = (8, 6), savefig = True, nickname = '', imgtype = 'svg'):
    
    igv_address = contig_string
    #Decompose the IGV address into component parts
    contig = igv_address.split(':')[0]
    start = igv_address.split(':')[1].split('-')[0].replace(',', '')
    end = igv_address.split(':')[1].split('-')[1].replace(',', '')
    igv_address_stripped = igv_address.replace(',', '').replace(':', '_').replace('-', '_')
    
    ####################################################
    ####################################################
    
    #Make a dummy region of interest bed file
    region = contig + '\t' + start + '\t' + end
    region_name = contig + '_' + start + '_' + end
    region_bed = 'dummy_region.bed'
    with open(region_bed, 'w+') as f:
        f.write(region)
    
    ImpulseParams_all = pd.DataFrame(pd.read_csv(allparamsloc))
    multicov = pd.DataFrame(pd.read_csv(multicovloc, sep = '\t'))
    
    #Run this block to get the fasta file of peaks overlapping your region of interest
    region_peaks_bed = 'dummy_region_peaks.bed'
    !/usr/local/bin/bedtools intersect -wa -a $peaksloc  -b $region_bed > $region_peaks_bed
    peaks = pd.read_csv(region_peaks_bed, sep = '\t', header = None)
    peaks = peaks[peaks[3] >= acc_filter]
    peaks[5] = peaks[0] + "_" + peaks[1].astype('str') + "_" + peaks[2].astype('str')
    #display(peaks)
    peaks_temp = peaks[5].values
    peaks_list = [i for i in peaks_temp if i in ImpulseParams_all['contig_start_end'].values]
    
    fig, ax = plt.subplots(figsize=figsize)
    
    for region in peaks_list:
        #Calculate fit given Impulse function expected ranges
        x_impulse_linspace = np.linspace(72, 192, 300)
        x_impulse = [72, 77, 80, 87, 90, 96, 104, 112, 120, 132, 144, 155, 168, 180, 192]
        x_impulse = [i for i in x_impulse for _ in (0, 1)]
    
        #Plot lines
        if alldata:
            impulse_params_for_contig_all = ImpulseParams_all.loc[ImpulseParams_all['contig_start_end'] == region].drop(columns = ['Unnamed: 0']).values.tolist()[0]
            impulse_fit_all = ImpulseEvaluate(x_impulse_linspace, impulse_params_for_contig_all)
            plt.plot(x_impulse_linspace, np.log(impulse_fit_all), alpha = 0.5, label = region)
        if scatter:
            all_multicov = multicov.loc[multicov['contig_start_end'] == region].drop(columns = ['contig_start_end']).astype('float').values.tolist()[0]
            plt.scatter(x_impulse, np.log(all_multicov), alpha = 0.5)

    plt.title(nickname + ' (' + igv_address + ') \n peak strength:' + str(acc_filter) + ' or higher')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.ylabel('log accessibility model fit')
    plt.xlabel('developmental stage')
    plt.yticks(fontsize = 12)
    plt.xticks([72, 77, 80, 87, 90, 96, 104, 112, 120, 132, 144, 155, 168, 180, 192], 
               ['S13', 'S14', 'S15', 'S17', 'S18', 'S19', 'S19+', 'S20', 'S21', 'S22', 'S23', 'S24', 'S25', 'S26', 'S27'], 
               rotation = 'vertical', fontsize = 12)
    
    ax2range = [i for i in range(60, 220, 20)]
    ax2 = plt.twiny()
    ax2.set_xticks([(i - 60) / (200 - 60) for i in ax2range])
    ax2.set_xticklabels(ax2range)
    ax2.set_xlabel('hours post fertilization')

    #contig = 'phaw_50.283864a_3320701_3323202'
    plot_name = nickname.replace(' ', '_') + '_' + igv_address_stripped if nickname != '' else igv_address_stripped
    file_name = plot_name + '.' + imgtype
    
    if savefig:
        plt.savefig(file_name)
    plt.show()
    
    return None
```

# Concordance vs. discordance analysis

```python
peaktypes_nearestgeneloc = '~/Labwork/Bioinformatics/Transcripts/' + 'Sall_peaktypes_nearestgene_new.gff'
genes_TMMexprloc = '~/Labwork/Bioinformatics/Parhyale_RNA-Seq/SLURM_scripts/' + 'kallisto.gene.TMM.EXPR.matrix'
Omni_multicov_mikadoloc = '~/Labwork/Bioinformatics/Omni-ATAC-Seq/OmniATAC_bothruns_peaks/' + 'OmniATAC_bothruns_multicov_unformat_mikado.tsv'
Omni_multicov_norm_mikadoloc = '~/Labwork/Bioinformatics/Omni-ATAC-Seq/OmniATAC_bothruns_peaks/' + 'OmniATAC_bothruns_multicov_mfuzz_mikado_standardised.tsv'
compfolder = '~/Labwork/Bioinformatics/Omni-ATAC-Seq/OmnivsRNA/'

def comparifier(peak_id, gene_id, compfolder, peaktypes_nearestgeneloc):
    if ':' in peak_id:
        peak_id = igv_to_peak(peak_id)
    
    comparisons = ['S19_vs_S13', 'S21_vs_S19', 'S23_vs_S21']
    
    peaktypes_nearestgene = pd.DataFrame(pd.read_csv(peaktypes_nearestgeneloc, sep = '\t', header = None))
    peaktypes_nearestgene['newstart'] = peaktypes_nearestgene[3] - 1
    peaktypes_nearestgene['contig_start_end'] = peaktypes_nearestgene[0].astype(str) + '_' + peaktypes_nearestgene['newstart'].astype('str') + '_' + peaktypes_nearestgene[4].astype('str')
    
    output_dict = {}

    for comparison in comparisons:
        RNAdfloc = compfolder + comparison + '_RNASeq_results.csv'
        RNAdf = pd.DataFrame(pd.read_csv(RNAdfloc))
        
        geneOI = RNAdf[RNAdf['Unnamed: 0'] == gene_id]
        
        Omnidfloc = compfolder + comparison + '_OmniATACSeq_results.csv'
        Omnidf = pd.DataFrame(pd.read_csv(Omnidfloc))
        
        peakOI = Omnidf[Omnidf['Unnamed: 0'] == peak_id]
    
        output_dict[comparison] = {
            'l2fc_acc': peakOI['log2FoldChange'].values[0],
            'padj_acc': peakOI['padj'].values[0],
            'l2fc_expr': geneOI['log2FoldChange'].values[0],
            'padj_expr': geneOI['padj'].values[0],
        }    
    
    return output_dict

def plot_expr_vs_acc(contig_string, peaktypes_nearestgeneloc, genes_TMMexprloc, Omni_multicov_mikadoloc, compfolder,
                    plot_expr = True, plot_acc = True, scatter = True, plot_cd = True, padj_cutoff = 0.05,
                    nickname = '', figsize = (5, 3), imgtype = 'svg', savefig = False):
    contig = igv_to_peak(contig_string)

    peaktypes_nearestgene = pd.DataFrame(pd.read_csv(peaktypes_nearestgeneloc, sep = '\t', header = None))
    peaktypes_nearestgene['newstart'] = peaktypes_nearestgene[3] - 1
    peaktypes_nearestgene['contig_start_end'] = peaktypes_nearestgene[0].astype(str) + '_' + peaktypes_nearestgene['newstart'].astype('str') + '_' + peaktypes_nearestgene[4].astype('str')

    stages = ['S13', 'S19', 'S21', 'S23']
    
    genes_TMMexpr = pd.DataFrame(pd.read_csv(genes_TMMexprloc, sep = '\t'))
    genes_TMMexpr.columns = ['gene', 'S13A', 'S13B', 'S13C', 'S19A', 'S19B', 'S19C', 'S21A', 'S21B', 'S21C', 'S23A', 'S23B', 'S23C']
    
    for stage in stages:
        stage_cols = [col for col in genes_TMMexpr.columns if stage in col]
        genes_TMMexpr[stage + '_mean_expr'] = genes_TMMexpr[stage_cols].mean(axis = 1)
    
    Omni_multicov_mikado = pd.DataFrame(pd.read_csv(Omni_multicov_mikadoloc, sep = '\t'))
    
    for stage in stages:
        stage_cols = [col for col in Omni_multicov_mikado.columns if stage in col]
        Omni_multicov_mikado[stage + '_mean_acc'] = Omni_multicov_mikado[stage_cols].mean(axis = 1)
    
    peaks_initial = Omni_multicov_mikado[['GENE.ID'] + [col for col in Omni_multicov_mikado.columns if 'mean' in col] + [stage + i for stage in stages for i in ['A', 'B']]].merge(peaktypes_nearestgene[['contig_start_end', 1, 9]], left_on = 'GENE.ID', right_on = 'contig_start_end')
    peaks_withgenes = peaks_initial[~peaks_initial[9].isna()]
    
    acc_names = [(i,i+'_acc') for i in [stage + i for stage in stages for i in ['A', 'B']]]
    peaks_withgenes.rename(columns = dict(acc_names), inplace=True)
    
    peaks_withall = peaks_withgenes.merge(genes_TMMexpr[['gene'] + [col for col in genes_TMMexpr.columns if 'mean' in col] + [stage + i for stage in stages for i in ['A', 'B', 'C']]], left_on = 9, right_on = 'gene')
    peaks_withall['expr_total'] = peaks_withall[[col for col in peaks_withall.columns if 'mean_expr' in str(col)]].sum(axis = 'columns')
    
    expr_names = [(i,i+'_expr') for i in [stage + i for stage in stages for i in ['A', 'B', 'C']]]
    peaks_withall.rename(columns = dict(expr_names), inplace=True)
    peak_indata = peaks_initial[peaks_initial['GENE.ID'] == contig]
    peak_ofinterest = peaks_withall[peaks_withall['GENE.ID'] == contig]

    times = [1, 2, 3, 4]
    times_acc = [i for i in times for j in np.arange(0, 2)]
    times_expr = [i for i in times for j in np.arange(0, 3)]
    
    if peak_indata.empty:
        print("The peak " + contig + " was not found. Double check your IGV address to make sure it matches a real peak.")
        return peaks_initial
    elif peak_indata[1].values[0] == 'unknown':
        print("The peak " + contig + " does not have a nearest gene.")
        return peak_indata
    elif peak_ofinterest['expr_total'].values[0] == 0:
        closest_gene = peak_ofinterest['gene'].values[0]
        print("The closest gene (" + closest_gene + ") near peak " + contig + " is not expressed in the RNA-Seq data.")
        
        acc_stagecols = [j + k + '_acc' for j in stages for k in ['A', 'B']]
        acc_values = [peak_ofinterest[i].values[0] for i in acc_stagecols]
        
        acc_mean_stagecols = [j + '_mean_acc' for j in stages]
        acc_averages = [peak_ofinterest[i].values[0] for i in acc_mean_stagecols]
        
        plt.figure(figsize = figsize)
        
        if plot_acc:
            plt.plot(times, acc_averages, color = 'tomato')
        
        if scatter:
            plt.scatter(times_acc, acc_values, color = 'tomato')
        
        plt.xticks(times, stages)
        plt.ylabel('Omni-ATAC accessibility')
        plt.xlabel('developmental stage')
        
        if nickname != '':
            plt.suptitle('Accessibility for ' + contig)
        else:
            plt.suptitle('Accessibility for ' + nickname + '\n(' + contig + ')')
        
        if savefig:
            plt.savefig('AccVExpr_' + contig + '.' + imgtype)
        plt.show()
        
        return peak_ofinterest
    else:
        closest_gene = peak_ofinterest['gene'].values[0]
        acc_stagecols = [j + k + '_acc' for j in stages for k in ['A', 'B']]
        acc_values = [peak_ofinterest[i].values[0] for i in acc_stagecols]
        
        acc_mean_stagecols = [j + '_mean_acc' for j in stages]
        acc_averages = [peak_ofinterest[i].values[0] for i in acc_mean_stagecols]
        
        expr_stagecols = [j + k + '_expr' for j in stages for k in ['A', 'B', 'C']]
        expr_values = [peak_ofinterest[i].values[0] for i in expr_stagecols]
        
        expr_mean_stagecols = [j + '_mean_expr' for j in stages]
        expr_averages = [peak_ofinterest[i].values[0] for i in expr_mean_stagecols]
        
        fig = plt.figure(figsize = figsize)
        
        gs = mpl.gridspec.GridSpec(3, 1, height_ratios=[2, 2, 2])
        ax1 = fig.add_subplot(gs[0, 0])
        
        if plot_acc:
            plt.plot(times, acc_averages, color = '#AF3034')
        
        if scatter:
            plt.scatter(times_acc, acc_values, color = '#AF3034')
        
        plt.xticks(times, stages)
        plt.ylabel('Omni-ATAC accessibility')
        plt.xlabel('developmental stage')
        
        if nickname != '':
            plt.title('Accessibility for ' + nickname + '\n(' + contig + ')')
        else:
            plt.title('Accessibility for ' + contig )
        
        ax2 = fig.add_subplot(gs[1, 0])
        
        if plot_cd:
            comparifier_out = comparifier(contig, closest_gene, compfolder, peaktypes_nearestgeneloc)
            
            l2fc_accs = [np.round(comparifier_out[i]['l2fc_acc'], 2) for i in comparifier_out]
            padj_accs = [comparifier_out[i]['padj_acc'] for i in comparifier_out]
            sig_accs = [str(l2fc_accs[i]) + '*' if padj_accs[i] < padj_cutoff else str(l2fc_accs[i]) for i in np.arange(0, len(l2fc_accs))]
            
            l2fc_exprs = [np.round(comparifier_out[i]['l2fc_expr'], 2) for i in comparifier_out]
            padj_exprs = [comparifier_out[i]['padj_expr'] for i in comparifier_out]
            sig_exprs = [str(l2fc_exprs[i]) + '*' if padj_exprs[i] < padj_cutoff else str(l2fc_exprs[i]) for i in np.arange(0, len(l2fc_exprs))]
            
            conc_disc = ['tomato' if l2fc_accs[i] * l2fc_exprs[i] > 0 else 'cornflowerblue' for i in np.arange(0, len(l2fc_exprs))]
            
            x = np.arange(1.5, 4, 1)  # the label locations
            width = 0.35  # the width of the bars

            rects1 = ax2.bar(x - width/2, l2fc_accs, width, label='accessibility log2 fold change', color = '#AF3034')
            rects2 = ax2.bar(x + width/2, l2fc_exprs, width, label='expression log2 fold change', color = '#9596C5')
            
            ax2.spines['top'].set_visible(False)
            ax2.spines['right'].set_visible(False)
            ax2.spines['bottom'].set_visible(False)
            ax2.spines['left'].set_visible(False)
            
            ax2.bar_label(rects1, sig_accs, padding=3)
            ax2.bar_label(rects2, sig_exprs, padding=3)
            
            ax2.tick_params(axis='both', which='both',length=0, pad = 10)
            plt.ylabel('log2 fold change')
            ylims = ax2.get_ylim()
            
            plt.hlines(0, 1, 4, color = 'black', linewidth = 0.5)
            
            plt.xticks(x, [i.replace('_', ' ') for i in comparifier_out.keys()])
            
            for m, color in enumerate(conc_disc):
                plt.gca().get_xticklabels()[m].set_color(color)
        
        ax3 = fig.add_subplot(gs[2, 0])
        
        if plot_expr:
            plt.plot(times, expr_averages, color = '#9596C5')
        
        if scatter:
            plt.scatter(times_expr, expr_values, color = '#9596C5')
        
        plt.xticks(times, stages)
        plt.ylabel('RNA-Seq kallisto TMM expression')
        plt.xlabel('developmental stage')

        plt.title('expression of ' + closest_gene)
        
        plt.suptitle('Accessibility vs. Expression')
        plt.tight_layout()
        
        if savefig:
            plt.savefig('AccVExpr_' + contig + '-vs-' + closest_gene + '.' + imgtype)
        plt.show()
        
        return peak_ofinterest

contig_string = 'phaw_50.283864a:3320702-3323202'
#contig_string = 'phaw_50.000014:238985-239410'
nickname = 'En1_pro'

peak_ofinterest = plot_expr_vs_acc(contig_string, peaktypes_nearestgeneloc, genes_TMMexprloc, Omni_multicov_mikadoloc, compfolder,
                plot_expr = True, plot_acc = True, scatter = True, plot_cd = True, padj_cutoff = 0.05,
                nickname = nickname, figsize = (4, 8), imgtype = 'svg', savefig = False)
display(peak_ofinterest)
```
