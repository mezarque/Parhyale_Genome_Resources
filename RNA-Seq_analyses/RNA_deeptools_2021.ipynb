```python
#############################################################
### Fill in this information before running this section ####

#Decide if you want to make a sample file
make_bw = False
run_refpoint = False
run_refpoint_TSS = False
run_scaled = False
run_refpoint_compared = True
run_scaled_compared = True

regionprefix = 'mikado_mRNA'
inbam = 'Phaw_RNA_hisat2.cleaned.sorted.bam'
chromsizes = 'phaw_5.0.chrom.sizes'

bgname = 'Phaw_RNA_hisat2.bedgraph'
bwname = 'Phaw_RNA_hisat2.bw'

bwname = 'Phaw_RNA_hisat2_bgsplit.bw'

regions_array = ['phaw_5.0.eggnog.resorted.mRNA.bed', 'mikado.loci.sorted.mRNAs.bed', 'mikado.loci.sorted.mRNAs.bed', 'Sall_promoters.bed']
comparison_array = ['phaw_5.0.eggnog.resorted.mRNA.bed', 'mikado.loci.sorted.mRNAs.bed']

#############################################################
#############################################################

if make_bw == True:
    !echo '### converting '{inbam}' to '{bgname}'\n'
    !bedtools genomecov -ibam {inbam} -g {chromsizes} -bg > {bgname}
    !echo '### converting '{bgname}' to '{bwname}'\n'
    !./bedGraphToBigWig {bgname} {chromsizes} {bwname}
        
if run_refpoint == True:
    for region in regions_array:
        refpoint_name = bwname + '_vs_' + region + '.deeptools.refpoint.gz'
        !echo '### computing refpoint matrix for '{refpoint_name}'\n'
        !computeMatrix reference-point -S {bwname} --regionsFileName {region} --outFileName {refpoint_name} -b 3000 -a 3000 --numberOfProcessors max --skipZeros --missingDataAsZero
        
        refpoint_pdf = refpoint_name.replace('.gz', '.pdf')
        !echo '### drawing figure into '{refpoint_pdf}'\n'
        !plotHeatmap --matrixFile {refpoint_name} --outFileName {refpoint_pdf}

if run_refpoint_TSS == True:
    for region in regions_array:
        refpoint_name = bwname + '_vs_' + region + '.deeptools.refpointTSS.gz'
        !echo '### computing refpoint matrix for '{refpoint_name}'\n'
        !computeMatrix reference-point -S {bwname} --regionsFileName {region} --outFileName {refpoint_name} --referencePoint=TSS -b 3000 -a 3000 --numberOfProcessors max --skipZeros --missingDataAsZero
        
        refpoint_pdf = refpoint_name.replace('.gz', '.pdf')
        !echo '### drawing figure into '{refpoint_pdf}'\n'
        !plotHeatmap --matrixFile {refpoint_name} --outFileName {refpoint_pdf}

if run_scaled == True:
    for region in regions_array:
        scaled_name = bwname + '_vs_' + region + '.deeptools.scaleregions.gz'
        !echo '### computing refpoint matrix for '{scaled_name}'\n'
        !computeMatrix scale-regions -S {bwname} --regionsFileName {region} --outFileName {scaled_name} -b 3000 -a 3000 --numberOfProcessors max --skipZeros --missingDataAsZero
        
        scaled_pdf = scaled_name.replace('.gz', '.pdf')
        !echo '### drawing figure into '{scaled_pdf}'\n'
        !plotHeatmap --matrixFile {scaled_name} --outFileName {scaled_pdf}

if run_refpoint_compared == True:
    comparision_string = ' '.join(comparison_array)
    refpoint_name = 'mikado_vs_Blondel_mRNAs.deeptools.refpointTSS_comparison_bgsplit.gz'
    !echo '### computing refpoint matrix for '{refpoint_name}'\n'
    !computeMatrix reference-point -S {bwname} --regionsFileName {comparision_string} --outFileName {refpoint_name} --referencePoint=TSS -b 3000 -a 3000 --numberOfProcessors max --skipZeros --missingDataAsZero
    
    refpoint_pdf = refpoint_name.replace('.gz', '.pdf')
    !echo '### drawing figure into '{refpoint_pdf}'\n'
    !plotHeatmap --matrixFile {refpoint_name} --outFileName {refpoint_pdf}

if run_scaled_compared == True:
    comparision_string = ' '.join(comparison_array)
    scaled_name = 'mikado_vs_Blondel_mRNAs.deeptools.scaleregions_comparison_bgsplit.gz'
    !echo '### computing refpoint matrix for '{scaled_name}'\n'
    !computeMatrix scale-regions -S {bwname} --regionsFileName {comparision_string} --outFileName {scaled_name} -b 3000 -a 3000 --numberOfProcessors max --skipZeros --missingDataAsZero
    
    scaled_pdf = scaled_name.replace('.gz', '.pdf')
    !echo '### drawing figure into '{scaled_pdf}'\n'
    !plotHeatmap --matrixFile {scaled_name} --outFileName {scaled_pdf}
```
