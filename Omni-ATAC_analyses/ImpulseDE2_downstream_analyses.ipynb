```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ImpulseDE_results = pd.DataFrame(pd.read_csv('OmniATAC-Seq_ImpulseDE2_allstages.csv'))

ImpulseDE_sorted = ImpulseDE_results.sort_values("padj")
#ImpulseDE_sorted.to_csv("OmniATAC-Seq_ImpulseDE2_allstages_sorted.csv")
```

```python
ImpulseDE_sorted.head(30)
```

```python
ImpulseDE2_flattened_modelfits = pd.DataFrame(pd.read_csv('OmniATAC-Seq_ImpulseDE2_allstages_flattened_modelfits.csv'))
ImpulseDE2_flattened_modelfits = ImpulseDE2_flattened_modelfits.add_suffix('_flat')

ImpulseDE2_allstages_modelfits = pd.DataFrame(pd.read_csv('OmniATAC-Seq_ImpulseDE2_allstages_modelfits.csv'))
ImpulseDE2_allstages_modelfits = ImpulseDE2_allstages_modelfits.add_suffix('_all')

ImpulseDE2_noS21S22_modelfits = pd.DataFrame(pd.read_csv('OmniATAC-Seq_ImpulseDE2_noS21S22_modelfits.csv'))
ImpulseDE2_noS21S22_modelfits = ImpulseDE2_noS21S22_modelfits.add_suffix('_noS21S22')
```

```python
ImpulseDE_sorted = ImpulseDE_sorted.merge(ImpulseDE2_flattened_modelfits, left_on = "Gene", right_on = "region_names_flat")

ImpulseDE_sorted = ImpulseDE_sorted.merge(ImpulseDE2_noS21S22_modelfits, left_on = "Gene", right_on = "region_names_noS21S22")

ImpulseDE_sorted = ImpulseDE_sorted.merge(ImpulseDE2_allstages_modelfits, left_on = "Gene", right_on = "region_names_all")
```

```python
ImpulseDE_significant = ImpulseDE_sorted[ImpulseDE_sorted["padj"] < 0.01]

ImpulseDE_significant.hist(column = ["maximum_flat", "maximum_noS21S22", "maximum_all"], bins = 15)
```

```python
ImpulseDE_significant["max_diff_all_vs_S21S22"] = abs(ImpulseDE_significant["maximum_noS21S22"] - ImpulseDE_significant["maximum_all"])

ImpulseDE_significant.hist(column = "max_diff_all_vs_S21S22", bins = 15)

ImpulseDE_significant["max_diff_all_vs_flat"] = abs(ImpulseDE_significant["maximum_flat"] - ImpulseDE_significant["maximum_all"])

ImpulseDE_significant.hist(column = "max_diff_all_vs_flat", bins = 15)
```

```python
nearest_neighbors = pd.DataFrame(pd.read_csv("OmniATAC_peaks/OmniATAC_q005_allpeaks.nearest_mRNA_cleaner.tsv", sep = '\t', header = None))
nearest_neighbors_withentry = nearest_neighbors[nearest_neighbors[10] != "."]

nearest_neighbors_withentry
```

```python
def get_last(string):
    dummy_array = string.split(';')
    if 'eggNOG-name' in dummy_array[-1].split('='):
        return dummy_array[-1].replace('eggNOG-name=', '')
    else:
        return dummy_array[-1]

nearest_annotation = [get_last(item) for item in nearest_neighbors[13]]

nearest_neighbors['Region'] = nearest_neighbors[0].astype(str) + "_" + nearest_neighbors[1].astype(str) + "_" + nearest_neighbors[2].astype(str)

nearest_neighbors['Nearby_Gene'] = nearest_annotation
nearest_neighbors

ImpulseDE_significant = ImpulseDE_significant.merge(right = nearest_neighbors, left_on = "Gene", right_on = "Region")

ImpulseDE_significant
```

```python
ImpulseDE_stable = ImpulseDE_significant[ImpulseDE_significant["max_diff_all_vs_S21S22"] < 6]

#ImpulseDE_stable.to_csv("OmniATAC-Seq_ImpulseDE2_stablediffregions_cleaner_withannotation.csv")
```

```python
ImpulseDE_stable
```

```python
ImpulseDE_first = ImpulseDE_significant[ImpulseDE_significant["maximum_all"] <= 6]

ImpulseDE_second = ImpulseDE_significant[ImpulseDE_significant["maximum_all"] <= 12]
ImpulseDE_second = ImpulseDE_second[ImpulseDE_second["maximum_all"] >= 7]

ImpulseDE_third = ImpulseDE_significant[ImpulseDE_significant["maximum_all"] <= 18]
ImpulseDE_third = ImpulseDE_third[ImpulseDE_third["maximum_all"] >= 13]

ImpulseDE_fourth = ImpulseDE_significant[ImpulseDE_significant["maximum_all"] <= 24]
ImpulseDE_fourth = ImpulseDE_fourth[ImpulseDE_fourth["maximum_all"] >= 19]

ImpulseDE_fifth = ImpulseDE_significant[ImpulseDE_significant["maximum_all"] >= 25]
```

```python
ImpulseDE_first_stable = ImpulseDE_first[ImpulseDE_first["max_diff_all_vs_S21S22"] < 6]
ImpulseDE_second_stable = ImpulseDE_second[ImpulseDE_second["max_diff_all_vs_S21S22"] < 6]
ImpulseDE_third_stable = ImpulseDE_third[ImpulseDE_third["max_diff_all_vs_S21S22"] < 6]
ImpulseDE_fourth_stable = ImpulseDE_fourth[ImpulseDE_fourth["max_diff_all_vs_S21S22"] < 6]
ImpulseDE_fifth_stable = ImpulseDE_fifth[ImpulseDE_fifth["max_diff_all_vs_S21S22"] < 6]
```
